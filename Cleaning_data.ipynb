```python
# Check any white space or empty vlaues are present in the data set or not.

object_cols = df.select_dtypes(include='object')

whitespace_issues = object_cols.apply(lambda col: col.apply(lambda x: isinstance(x, str) and x.strip() == '')).sum()

columns_with_whitespace = whitespace_issues[whitespace_issues > 0]

print("Columns with empty or whitespace-only values:")
print(columns_with_whitespace)


```

-> there is no column with empty value or white space 

```python
# Strip and lowercase all object columns to make sure all column are clean, consistent, and uniform.

for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip().str.lower()
```

```python
df.head()
```

```python
# Use IQR to remove outliers
Q1 = df['Average Cost for two'].quantile(0.25)
Q3 = df['Average Cost for two'].quantile(0.75)
IQR = Q3 - Q1

print(IQR)
```

```python
# Step 2: Define boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Calculate median 
valid_data = df[(df['Average Cost for two'] >= lower_bound) & (df['Average Cost for two'] <= upper_bound)]
median_value = valid_data['Average Cost for two'].median()

# Replace outliers with median
df['Average Cost for two'] = df['Average Cost for two'].apply(
    lambda x: median_value if x < lower_bound or x > upper_bound else x
)
```

```python
sns.boxplot(x=df['Average Cost for two'])
plt.title("Boxplot Average Cost for Two (After Outlier Removal)")
plt.show()
```

-> The boxplot helps visualize the distribution of restaurant pricing for two people.

-> **Most restaurants** charge between **₹250 and ₹1000** for two people.

-> The **median (middle)** cost lies around **₹500**.

-> A few **outliers** still exist around ₹1200–₹1350 (marked as small circles).

-> The **majority of data is clean** and tightly packed within a reasonable price range.

-> Outlier removal helped reduce extreme distortions (like ₹800,000 seen earlier).
